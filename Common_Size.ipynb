```python
# Helper Function for finding index of string value in a column
def check_index(column,val):
    a = df.index[df[column].str.contains(val)]
    if a.empty:
        return 'not found'
    elif len(a) > 1:
        return a.tolist()
    else:
        #only one value - return scalar
        return a.item()


```

```python
SABS = check_index('Unnamed: 0','Standardized Annual Balance Sheet')
print(SABS)
TLE = check_index('Unnamed: 0','Total Liabilities & Equity')[0]
print(TLE)
SAIS = check_index('Unnamed: 0','Standardized Annual Income Statement')
print(SAIS)
SO = check_index('Unnamed: 0','Shares Outstanding')
print(SO)
SARE = check_index('Unnamed: 0','Standardized Annual Retained Earnings')
print(SARE)
NPPE = check_index('Unnamed: 0','Net Purch of Pty Plant & Equip')
print(NPPE)
```

```python

# Load the workbook
wb = openpyxl.load_workbook(new_path)

ws = wb['Sheet1']
# Rename the worksheet
ws.title = 'Main'

# Load the main sheet
main_sheet = wb['Main']

# # Create a new sheet
# balance_sheet = wb.create_sheet("Balance Sheet")

# Save the changes to the workbook
wb.save(new_path)
```

```python
# Load the workbook
wb = openpyxl.load_workbook(new_path)

# Load the main sheet
main_sheet = wb['Main']

# Create a new sheet
balance_sheet = wb.create_sheet("Balance Sheet")

# Copy data from main sheet to balance sheet
for i in range(SABS+3, TLE+3):
    for j in range(1, main_sheet.max_column+1):
        balance_sheet.cell(row=i-2, column=j).value = main_sheet.cell(row=i, column=j).value

# Save the changes to the workbook
wb.save(new_path)
```

```python
wb = openpyxl.load_workbook(new_path)

# Load the main sheet
main_sheet = wb['Main']

# Create a new sheet
Income_sheet = wb.create_sheet("Income Sheet")

# Copy data from main sheet to Income sheet
for i in range(SAIS+3, SO+3):
    for j in range(1, main_sheet.max_column+1):
        Income_sheet.cell(row=i-SAIS-2, column=j).value = main_sheet.cell(row=i, column=j).value

# Save the changes to the workbook
wb.save(new_path)
```

```python
wb = openpyxl.load_workbook(new_path)

# Load the main sheet
main_sheet = wb['Main']

# Create a new sheet
Retained_Earnings_sheet = wb.create_sheet("Retained Earnings Sheet")

# Copy data from main sheet to Retained Earnings sheet
for i in range(SARE+3, NPPE+3):
    for j in range(1, main_sheet.max_column+1):
        Retained_Earnings_sheet.cell(row=i-SARE-2, column=j).value = main_sheet.cell(row=i, column=j).value

# Save the changes to the workbook
wb.save(new_path)
```
